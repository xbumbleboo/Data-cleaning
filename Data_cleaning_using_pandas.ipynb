```python
import pandas as pd
```

```python
df = pd.read_csv('layoffs.csv')
```

```python
df
```

```python
df.info()
```

```python
duplicate_rows = df[df.duplicated()]
len(duplicate_rows)
duplicate_rows
```

```python
df.drop_duplicates(inplace=True)
```

```python
duplicate_rows = df[df.duplicated()]
len(duplicate_rows)
```

```python
df['funds_raised_millions'].describe()
```

```python
negative_values = df[df['funds_raised_millions'] < 0]
len(negative_values)
```

```python
df = df[df['funds_raised_millions'] >= 0]
negative_values = df[df['funds_raised_millions'] < 0]
negative_values
```

```python
df.isnull().sum()
```

```python
numerical_columns_with_null = df.select_dtypes(include=['float64', 'int64']).columns
for col in ['total_laid_off', 'percentage_laid_off']:
    if col in numerical_columns_with_null:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df[df['stage'].notna()]
```

```python
df.isnull().sum()
```

```python
df = df[df['industry'].notna()]
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y')
df.info()
```

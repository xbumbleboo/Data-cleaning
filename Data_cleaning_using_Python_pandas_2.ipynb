```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

# 1. Phát hiện vấn đề  
Đọc bảng và xem xét giá trị trong các cột  
- Có vài ký thự không hợp lệ  
- Vài giá trị NULL  
- Cột 'Phone_Number' không thống nhất định dạng  
- Thống nhất dữ liệu địa chỉ, Paying_customer và Do_not_contact

# 2. Xóa duplicate  
Phát hiện duplicate xem có bao nhiêu duplicate

```python
duplicate_rows = df[df.duplicated()]
len(duplicate_rows)
```

Lúc đầu có 1 duplicate, sau khi dùng lệnh drop_duplicate chúng ta đã xóa hết dữ liệu trùng

```python
df = df.drop_duplicates()
df
```

# 3. Xóa cột không dùng đến  
Sử dụng drop(column)

```python
df = df.drop(columns = "Not_Useful_Column")
```

# 4. Ký tự lỗi/không phù hợp  
Sau khi quan sát cột 'Last_Name', ta thấy xuất hiện các ký tự trong tên như "/, ..., _" khiến tên hiển thị không đồng nhất. Ta cần xóa những ký tự này bằng cách sử dụng lstrip và rstrip cho dạng string.

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df

#Có một cách khác có thể xóa toàn bộ những ký tự này chỉ trong 1 dòng code, đó là kết hợp chúng vào 1 chuỗi giá trị như dưới đây chứ không phải 1 danh sách chuỗi
#df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

# 5. Thay đổi/thống nhất định dạng số điện thoại  
Định dạng số điện thoại ở bảng dữ liệu gốc đang có hỗn hợp rất nhiều số và kí tự vào với nhau. Việc của chúng ta là lấy ra tất cả những giá trị không phải số và xóa nó đi, sau đó đưa số điện thoại về chỉ số thôi rồi ép nó vào định dạng chúng ta muốn

## 5.1. Xóa những ký tự không phải số  
Chúng ta sẽ xóa bất kỳ ký tự nào không phải số ra khỏi cột Phone_Number và thay vào đó là khoảng trống/không gì cả

```python
df['Phone_Number'] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df

# Sau khi xử lý các ký tự không hợp lệ, chúng ta còn lại 1 chuỗi số như bên dưới
```

## 5.2. Định dạng lại số điện thoại

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# Sau khi viết lệnh bên trên, ta bị lỗi vì tất cả dãy số đều là float, không áp dụng được --> cần chuyển sang string

# sau khi thử thành công, ta áp dụng toàn bộ cột phone_number bằng định dạng string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# Khi này sdt đã thành string, chúng ta có thể áp dụng code fail lúc đầu
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# Phần cuối cùng là định dạng null/nan/NaN..., ta cần xóa đi vì không có cách nào để điền sdt
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

# 6. Chia cột address  
Addesss đang bao gồm những thông tin như Street, State, Zipcode. Chúng ta sẽ tách nó thành 3 cột riêng

```python
df[["Street", "State", "Zipcode"]]= df['Address'].str.split(',', expand=True)
```

```python
df
```

# 7. Thống nhất định dạng Yes/No 2 cột Paying_Customer và Do_not_contact  
Có thể thấy rằng Yes/No 2 cột này mang nhiều định dạng Yes/Y/No/N, ta sẽ đưa nó về đồng nhất Y và N

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer']
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

# 8. Xử lý giá trị NULL cả bảng  
Sau những quy trình bên trên, xót lại những định dạng NULL khác mà các bước chuẩn hóa từng cột không làm được. Ta sẽ xóa NULL trên toàn bộ bảng

```python
# Điền những ô giá trị null bằng không gì cả
df=df.fillna('')
df
```

```python
df.replace('N/a','')
df
```

# 9. Điền giá trị khuyết  
Với cột Do_not_contact, những khách hàng 'Y' là những khách hàng không mong muốn được liên hệ, 'N' là có, và chúng ta không muốn bỏ lỡ những khách hàng blank. Suy ra chúng ta cần xóa những hàng 'Y' và điền blank bằng 'N' để có 1 danh sách chỉ các khách hàng mong muốn được liên hệ

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df

#another way to drop null values
#df = df.dropna(subset='Phone_Number'), inplace = True
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('','N')
df
```

```python
# làm tương tự với cột phone_number, xóa những hàng không có sdt đồng nghĩa không thể liên hệ
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```
